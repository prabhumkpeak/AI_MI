Example for Pandas filter and sorting

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url , sep='\t')
chipo.head(5)
```

```python
new_price = [float(value[1:-1]) for value in chipo.item_price] #Convert price string to float
```

```python
chipo["item_price"] = new_price #Add new price column
```

Task 1 : Drop duplicate row with item_name, Quantity and choice_descrioption

```python
chipo.drop_duplicates(subset=['item_name', 'quantity', 'choice_description'], inplace=True) # Drop duplicate row with item_name, Quantity and choice_descrioption
```

Task 2: How many products crossed $10.00

```python
chipo.query('item_price > 10.00') # Count how many products crossed 10.00
```

```python
chipo[chipo['item_price'] > 10.00] # Count how many products crossed 10.00
```

```python
# filter item_name is Chicken Bowl and quantity is 1
chipo[(chipo['item_name'] == 'Chicken Bowl') & (chipo['quantity'] == 1)] # Filter the rows
```

```python
#select only products with quantity is 1
chipo[chipo['quantity'] == 1] # Filter the rows
```

```python
chipo[chipo['item_name'] == 'Chicken Bowl'] # Filter the rows
```

Task 3 : What is the price of each item?

```python
# display only two columns iteam_name and item_price
chipo.filter(items=['item_name', 'item_price'], axis=1) # Filter the columns
```

```python
# display only two columns iteam_name and item_price
chipo[['item_name', 'item_price']].nunique()# Filter the columns
```

Task 4 : sort the values from the most to less expensive

```python
chipo.sort_values(by='item_price', ascending=False) # sort the values from the most to less expensive
```

```python
#Sort by item_name
chipo.item_name.sort_values() # Sort the values from A to Z
```

```python
chipo.sort_values(by='item_name') # Sort the values from A to Z
```

Task 5 : What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by='item_price', ascending=False).head(1) # Sort the values from A to Z
```

Task 6 : How many times was a Veggie Salad Bowl ordered?

```python
chipo_salad = chipo[chipo.item_name == 'Veggie Salad Bowl'] # Filter the rows for Veggie Salad Bowl
len(chipo_salad)  # How many times was a Veggie Salad Bowl ordered?
```

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl'].quantity.sum() # How many times was a Veggie Salad Bowl ordered?
```

Task 7 : How many times did someone order more than one Canned Soda?

```python
chipo_soda = chipo[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1) ]# How many times did someone order more than one Canned Soda?
len(chipo_soda)  # How many times did someone order more than one Canned Soda?
```
