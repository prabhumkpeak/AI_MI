MPG car dataset for Merge execrise in PANDAS

```python
# import pandas and numpy
import pandas as pd
import numpy as np
```

```python
#import the MPG car dataset
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")

```

```python
cars1.head()
```

TASK 1 : Drop the unnamed columns in cars1 dataset


```python
cars1 = cars1.loc[:,'mpg':'car']
```

```python
cars1.head()
```

```python
cars2.head()
```

TASK 2 : Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1, cars2], ignore_index=True)
```

```python
cars
```

TASK 3 : there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
RM_owner = np.random.randint(15000, 73000, size=len(cars), dtype=int)
```

TASK 4 : Add the column owners to cars

```python
cars['owner'] = RM_owner
```
