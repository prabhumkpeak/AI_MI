Visualizing the Titanic Disaster

Load the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
url = "C:\\AI-ML\\AI_MI_Career\\datasets\\train.csv"
titanic = pd.read_csv(url)
titanic.head()
```

TASK 1 : Set PassengerId as the index

```python
titanic.set_index('PassengerId', inplace=True)
```

TASK 2 : Create a pie chart presenting the male/female proportion

```python
SEX_ratio = titanic.groupby('Sex').size()
SEX_ratio.head()
```

```python
plt.pie( SEX_ratio, labels=SEX_ratio.index, autopct='%1.1f%%', startangle=140 )
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('SEX RATIO')
plt.show
```

TASK 2 : Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
lm = sns.scatterplot(data=titanic, x='Age', y='Fare', hue = 'Sex', legend='full')
lm.set_title('Scatterplot of Age vs Fare')
```

TASK 3 : How many people survived?

```python
len(titanic[titanic['Survived'] == 1])
```

TASK 4 : Create a histogram with the Fare payed

```python
fare_hist = sns.histplot(titanic['Fare'], bins=30, kde=True)
fare_hist.set_title('Histogram of Fare Paid')
```
